```python
case_study_data = df[df['Article Title'] == "11 Best Solo Leveling Arcs in the Manhwa to Read Now"]
```

```python
# Print the relevant values before creating the user segmentation
print("Age Group 18-24 (%):", case_study_data["Age Group 18-24 (%)"].mean())
print("Age Group 25-34 (%):", case_study_data["Age Group 25-34 (%)"].mean())
print("Returning Visitors (%):", case_study_data["Returning Visitors (%)"].mean())
print("New Visitors (%):", case_study_data["New Visitors (%)"].mean())
```

```python
user_segmentation = {
    "Demographic Segment": ["18-24", "25-34", "Returning Visitors", "New Visitors"],
    "Percentage (%)": [
        case_study_data["Age Group 18-24 (%)"].median(),
        case_study_data["Age Group 25-34 (%)"].median(),
        case_study_data["Returning Visitors (%)"].median(),
        case_study_data["New Visitors (%)"].median(),
    ]
}
user_segmentation_df = pd.DataFrame(user_segmentation)
total = user_segmentation_df['Percentage (%)'].sum()
user_segmentation_df['Percentage (%)'] = (user_segmentation_df['Percentage (%)'] / total) * 100
```

```python
print(user_segmentation_df)
```

```python
print("Total Percentage:", user_segmentation_df['Percentage (%)'].sum())
```

```python
plt.figure(figsize=(4, 6))
plt.pie(user_segmentation_df['Percentage (%)'], labels=user_segmentation_df['Demographic Segment'], autopct='%1.1f%%', startangle=90)
plt.title('User Segmentation for Solo Leveling Arcs')
plt.axis('equal')
plt.show()
```
